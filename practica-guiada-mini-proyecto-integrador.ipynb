```python
import random  #importamos la librería random para generar números y elecciones al azar
```

```python
def generar_datos(n=100):  # n=100 es el valor por defecto, si no pones nada genera 100 préstamos
    
    # lista con las cuatro categorías posibles
    categorias = ["Ficción", "Ensayo", "Cómic", "Poesía"]
    
    # lista vacía donde iremos guardando cada préstamo
    prestamos = []
    
    # bucle que se repite n veces (100 por defecto)
    # range(n) genera los números 0, 1, 2... hasta n-1
    # i es el número de la vuelta actual
    for i in range(n):
        
        # creamos un diccionario con los datos de este préstamo
        prestamo = {
            "id": i + 1,                          # número del préstamo (empieza en 1, no en 0)
            "categoria": random.choice(categorias), # elige una categoría al azar de la lista
            "dias": random.randint(5, 45)           # número al azar entre 5 y 45 (ambos incluidos)
        }
        
        # añadimos este préstamo a la lista
        prestamos.append(prestamo)
    
    # devolvemos la lista completa con los 100 préstamos
    return prestamos

#comprobamos que se haya creado correctamente:
datos = generar_datos(100)
print(datos[:3])  
```

```python
def analizar_por_categoria(prestamos):
    # diccionario vacío donde agruparemos los días por categoría
    # al final quedará así: {"Ficción": [23, 12, 45...], "Cómic": [8, 33...], ...}
    por_categoria = {}

    # recorremos cada préstamo de la lista
    for prestamo in prestamos:
        cat = prestamo["categoria"]  # sacamos la categoría de este préstamo ("Ficción", "Cómic"...)
        dias = prestamo["dias"]      # sacamos los días de este préstamo (un número entre 5 y 45)

        # si esta categoría no existe aún en el diccionario, la creamos con una lista vacía
        # esto evita un error al intentar añadir a una lista que no existe
        if cat not in por_categoria:
            por_categoria[cat] = []

        # añadimos los días a la lista de esa categoría
        por_categoria[cat].append(dias)

    # calculamos estadísticas para cada categoría
    total = len(prestamos)  # total de préstamos (100)

    for cat, lista_dias in por_categoria.items():  # en cada vuelta: cat = nombre, lista_dias = sus días
        cantidad = len(lista_dias)                      # cuántos préstamos tiene esta categoría
        promedio = sum(lista_dias) / len(lista_dias)    # promedio de días de esta categoría
        porcentaje = (cantidad / total) * 100           # qué % del total representa

        print(f"  {cat}: {cantidad} préstamos | {promedio:.1f} días | {porcentaje:.1f}%")

datos = generar_datos(100)
print("Por categoría:")
analizar_por_categoria(datos)
```

```python
def analizar_por_estado(prestamos):
    conteo = {"A tiempo": 0, "Retraso leve": 0, "Retraso grave": 0}

    for prestamo in prestamos:
        dias = prestamo["dias"]

        if dias <= 21:       # 0 a 21 días
            conteo["A tiempo"] += 1
        elif dias <= 30:     # 22 a 30 días (ya sabemos que > 21)
            conteo["Retraso leve"] += 1
        else:                # más de 30 días
            conteo["Retraso grave"] += 1

    total = len(prestamos)
    for estado, cantidad in conteo.items():
        porcentaje = (cantidad / total) * 100
        print(f"  {estado}: {cantidad} préstamos ({porcentaje:.1f}%)")

datos = generar_datos(100)
analizar_por_estado(datos)
```

```python
def calcular_penalizaciones(prestamos):
    penalizacion_total = 0
    prestamos_con_retraso = 0

    for prestamo in prestamos:
        dias = prestamo["dias"]       # ✅ comillas

        if dias <= 21:
            penalizacion = 0
        elif dias <= 30:
            dias_extra = dias - 21        # ✅ desde el día 21
            penalizacion = dias_extra * 0.50
        else:
            dias_extra = dias - 21        # ✅ siempre desde el día 21
            penalizacion = dias_extra * 1.00

        if penalizacion > 0:
            penalizacion_total += penalizacion
            prestamos_con_retraso += 1    # ✅ suma 1

    promedio = penalizacion_total / prestamos_con_retraso
    print(f"  Total recaudado: {penalizacion_total:.2f}€")
    print(f"  Promedio por retraso: {promedio:.2f}€")

datos = generar_datos(100)
calcular_penalizaciones(datos)
```

```python
def top5_prestamos(prestamos):
    # ordenamos la lista de mayor a menor días
    ordenados = sorted(prestamos, key=lambda p: p["dias"], reverse=True)
    
    # cogemos solo los 5 primeros
    top5 = ordenados[:5]
    
    print("  Top 5 préstamos más largos:")
    for prestamo in top5:
        print(f"  ID {prestamo['id']} | {prestamo['categoria']} | {prestamo['dias']} días")

datos = generar_datos(100)
top5_prestamos(datos)
```
