**Procesar y mostrar estadísticas**   

*Pistas*
- La función debe devolver un diccionario con estadoypenalizacion

- Cuentas cuantas estan en cada categoria

```python
def clasificar_prestamo(dias):
    if dias <= 21:
        estado = "A tiempo"
        penalizacion = 0
    elif dias <= 30:
        dias_extra = dias - 21
        estado = "Retraso leve"
        penalizacion = dias_extra * 0.50
    else:
        dias_extra = dias - 21
        estado = "Retraso grave"
        penalizacion = dias_extra * 1.00

    return {"estado": estado, "penalizacion": penalizacion}


prestamos_dias = [15, 22, 18, 35, 25, 12, 40, 19, 28, 33]

conteo = {"A tiempo": 0, "Retraso leve": 0, "Retraso grave": 0}
penalizacion_total = 0

print("\033[1mClasificador de préstamos 📚\033[0m")
print("-" * 40)

for dias in prestamos_dias:
    resultado = clasificar_prestamo(dias)
    estado = resultado["estado"]
    penalizacion = resultado["penalizacion"]

    conteo[estado] += 1
    penalizacion_total += penalizacion

    if penalizacion > 0:
        print(f"  {dias} días → {estado} | {penalizacion:.2f}€")
    else:
        print(f"  {dias} días → {estado} ✅")

print("-" * 40)
total = len(prestamos_dias)
print(f"\n📊 Estadísticas:")
for categoria, cantidad in conteo.items():
    porcentaje = (cantidad / total) * 100
    print(f"  {categoria}: {cantidad} ({porcentaje:.1f}%)")

print(f"\n💶 Total penalizaciones: {penalizacion_total:.2f}€")
```


##### ✨ **Ejercicio 3: Análisis de categorías** ✨
**Objetivo: Practicar listas, diccionarios y bucles**

**Enunciado:** Tienes una lista de libros prestados con su categoría y días de préstamo:

prestamos = [    
    {"titulo": "1984", "categoria": "Ficción", "dias": 15},   
    {"titulo": "Sapiens", "categoria": "Ensayo", "dias": 22},   
    {"titulo": "Watchmen", "categoria": "Cómic", "dias": 18},   
    {"titulo": "El Quijote", "categoria": "Ficción", "dias": 30},   
    {"titulo": "Breve historia", "categoria": "Ensayo", "dias": 25},   
    {"titulo": "Batman", "categoria": "Cómic", "dias": 12}   
]

**Escribe un programa que:**   
- Agrupe los préstamos por categoría    
- Calcular el promedio de días por categoría    
- Identifica la categoría más prestada   
- Muestre libros con más de 20 días   

**Pistas**
- Crea un diccionario para agrupar por categoría.
- Usa un bucle para iterar sobre los préstamos
- Para cada categoría, guarda una lista de días.
- Calcula promedios usando sum()ylen()


```python
prestamos = [    
    {"titulo": "1984", "categoria": "Ficción", "dias": 15},   
    {"titulo": "Sapiens", "categoria": "Ensayo", "dias": 22},   
    {"titulo": "Watchmen", "categoria": "Cómic", "dias": 18},   
    {"titulo": "El Quijote", "categoria": "Ficción", "dias": 30},   
    {"titulo": "Breve historia", "categoria": "Ensayo", "dias": 25},   
    {"titulo": "Batman", "categoria": "Cómic", "dias": 12}   
]

#diccionario vacío
por_categoria = {}

# recorremos cada libro de la lista
for prestamo in prestamos:

    # sacamos la categoría y los días de este libro
    cat = prestamo["categoria"]
    dias = prestamo["dias"]

    # si esta categoría no existe aún en el diccionario, la creamos
    if cat not in por_categoria:
        por_categoria[cat] = []

    # añadimos los días a la lista de esa categoría
    por_categoria[cat].append(dias)

#promedio
for categoria, lista_dias in por_categoria.items():
    promedio = sum(lista_dias) / len(lista_dias)
    print(f"  {categoria}: promedio de {promedio:.1f} días")

#Categoria más prestada
categoria_max = max(por_categoria, key=lambda cat: len(por_categoria[cat]))
print(f" Categoría más prestada: {categoria_max}")

#en esta caso vemos claro que hay 2 por categoria, así que max devuelve la primera que encuentra... para saber si hay empate:

# primero averiguas cuál es el número máximo
max_libros = max(len(lista) for lista in por_categoria.values())

# luego buscas todas las categorías que tienen ese número
categorias_max = [cat for cat, lista in por_categoria.items() if len(lista) == max_libros]

print(f"  Categorías más prestadas: {categorias_max} con {max_libros} libros")


# Mostrar libros con más de 20 dias

print(f"\nLibros con más de 20 días:")
for prestamo in prestamos:
    if prestamo["dias"] > 20:
        print(f"  {prestamo['titulo']} — {prestamo['dias']} días")
```

def limpiar_datos(ventas):

def calcular_estadisticas(ventas):
    """Devuelve dict con total, promedio, max, min"""

def clasificar_ventas(ventas, umbral_bajo, umbral_alto):
    """Devuelve dict con ventas bajas, medias, altas"""

def generar_reporte(ventas):
    """Usa las funciones anteriores y muestra reporte completo"""

**Datos de prueba (con valores inválidos):**

*ventas_brutas = [1500, -200, 2500, None, 3000, 0, 4500, -100, 2800, 5000]*

**Requerimientos**
- limpiar_datos: Filtrar valores válidos (> 0 y no Ninguno)   
- calcular_estadisticas: Total, promedio, máximo, mínimo   
- clasificar_ventas:      
Baja: < umbral_bajo   
Media: entre umbrales   
Alta: >= umbral_alto   
- generar_reporte: Llamar todas las funciones y mostrar resultados formateados   
- Salida esperada (aproximada)   

=== REPORTE DE VENTAS ===

Datos procesados: 7 ventas válidas (3 inválidas eliminadas)

Estadísticas:
  Total: 21800
  Promedio: 3114.29
  Máximo: 5000
  Mínimo: 1500

Clasificación:
  Ventas bajas (< 2000): 1 (14.3%)
  Ventas medias: 4 (57.1%)
  Ventas altas (>= 4000): 2 (28.6%)

```python
ventas_brutas = [1500, -200, 2500, None, 3000, 0, 4500, -100, 2800, 5000]

def limpiar_datos(ventas):
    resultado = []
    for v in ventas:
        if v is not None and v > 0:
            resultado.append(v)
    return resultado

def calcular_estadisticas(ventas):
    return {
        "total":    sum(ventas),
        "promedio": sum(ventas) / len(ventas),
        "maximo":   max(ventas),
        "minimo":   min(ventas)
    }

def clasificar_ventas(ventas):
    bajas  = []
    medias = []
    altas  = []
    for v in ventas:
        if v < 2000:
            bajas.append(v)
        elif v >= 4000:
            altas.append(v)
        else:
            medias.append(v)
    return {"bajas": bajas, "medias": medias, "altas": altas}

def generar_reporte(ventas):
    limpias   = limpiar_datos(ventas)
    invalidas = len(ventas) - len(limpias)
    stats     = calcular_estadisticas(limpias)
    clases    = clasificar_ventas(limpias)
    total     = len(limpias)

    print("REPORTE DE VENTAS")
    print(f"Ventas válidas: {total} ({invalidas} eliminadas)\n")
    print(f"Total:    {stats['total']}")
    print(f"Promedio: {stats['promedio']:.2f}")
    print(f"Máximo:   {stats['maximo']}")
    print(f"Mínimo:   {stats['minimo']}\n")
    print(f"Bajas  (< 2000):  {len(clases['bajas'])}")
    print(f"Medias:           {len(clases['medias'])}")
    print(f"Altas  (>= 4000): {len(clases['altas'])}")

generar_reporte(ventas_brutas)
```
